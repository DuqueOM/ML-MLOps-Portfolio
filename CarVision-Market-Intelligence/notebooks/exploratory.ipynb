# CarVision Exploratory Analysis
This notebook performs basic EDA on vehicles_us.csv.

```python
import pandas as pd
import plotly.express as px
from pathlib import Path

DATA='..//vehicles_us.csv'
assert Path(DATA).exists(), 'Dataset not found'
df=pd.read_csv(DATA)
df.head()
```

```python
df.describe(include='all').T
```

```python
fig=px.histogram(df, x='price', nbins=50, title='Price Distribution')
fig.show()
```

```python
if 'odometer' in df.columns:
    fig=px.scatter(df.sample(min(2000,len(df))), x='odometer', y='price', opacity=0.4, title='Price vs Odometer')
    fig.show()
```
